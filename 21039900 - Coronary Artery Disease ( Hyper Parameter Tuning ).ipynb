```python
# Initialize variables to track the highest AUC score and corresponding iteration
max_auc_score = 0
best_iteration = 0

for i in range(20):
    # Randomly choose hyperparameters
    max_depth = random.choice(max_depth_values)
    min_samples_split = random.choice(min_samples_split_values)
    min_samples_leaf = random.choice(min_samples_leaf_values)

    # Create and train the Decision Tree model
    tree = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf)
    tree.fit(x_train, y_train)

    # Make predictions on test data
    y_pred = tree.predict(x_test)

    # Calculate AUC score
    fpr, tpr, thresholds = roc_curve(y_test, y_pred)
    auc_score = auc(fpr, tpr)

    # Check if the current iteration has a higher AUC score
    if auc_score > max_auc_score:
        max_auc_score = auc_score
        best_iteration = i

# Re-run the best iteration and plot its ROC curve
max_depth = random.choice(max_depth_values)
min_samples_split = random.choice(min_samples_split_values)
min_samples_leaf = random.choice(min_samples_leaf_values)

tree = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf)
tree.fit(x_train, y_train)

y_pred = tree.predict(x_test)

# Plot ROC curve for the best iteration
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, label=f'Best Iteration {best_iteration + 1} (AUC = {max_auc_score:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
plt.title(f'ROC Curve - Best Iteration {best_iteration + 1}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
